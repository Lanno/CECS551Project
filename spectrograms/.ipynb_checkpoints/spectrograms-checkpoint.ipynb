```python
from matplotlib import pyplot as plt

import medleydb as mdb

from scipy import signal
from scipy.io import wavfile
```

```python
dir(mdb)
```

```python
help(mdb.load_multitracks)
```

```python
tracks = mdb.load_multitracks(["Phoenix_ColliersDaughter"])
```

```python
track = next(tracks)
```

```python
help(track)
```

```python
track.audio_path
```

```python
track.mix_path
```

```python
dir(track)
```

```python
track.stem_instruments
```

```python
flutes = mdb.get_files_for_instrument("flute", [track])
```

```python
flute = next(flutes)
```

```python
help(wavfile.read)
```

```python
sample_rate, audio = wavfile.read(track.mix_path)
audio = audio.mean(1)
audio = audio[sample_rate*8:int(-sample_rate*1.5)]
```

```python
help(signal.spectrogram)
```

```python
freqs, times, sxx = signal.spectrogram(audio, fs=sample_rate, mode="magnitude")
```

```python
print(freqs.shape, times.shape, sxx.shape)
```

```python
help(plt.pcolormesh)
```

```python
plt.pcolormesh(times, freqs, sxx)
```

```python
help(plt.specgram)
```

```python
data, freqs, times, image = plt.specgram(audio, Fs=sample_rate, mode="magnitude", NFFT=512, noverlap=256)
```

```python
data.shape
```
